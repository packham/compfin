```python
import numpy as np
import pandas as pd
from pylab import mpl , plt
import statsmodels.api as sm
plt.style.use('seaborn')
```

```python
hprice = pd. read_excel ('./data/hprice.xls')
Y = hprice ['sale price ']
X = hprice ['# bedroom ']
X = sm. add_constant (X)
model = sm.OLS(Y,X)
results = model .fit ()
results . summary ()
```
